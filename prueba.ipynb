```python
import numpy as np
def punto_medio(a,b,f,N):
    return (b-a)/N * sum( f(a + (i + 0.5)*(b-a)/N) for i in range(N) ) 
```

```python
f = lambda x: x**2
punto_medio(0,10,f,50)
```

```python
def rectangulo(a,b,f,N):
    return (b-a)/N * sum( f(a + i*(b-a)/N) for i in range(N) )  
def trapecio(a,b,f,N):
    return (b-a)/N * ( (f(a) + f(b))/2 + sum( f(a + i*(b-a)/N) for i in range(1,N) ) )
```
